```python
import pandas as pd # Pandas     /   (DataFrames)
import matplotlib.pyplot as plt #    (figures, , ).
 
#   dataset
df = pd.read_excel("Thyroid_Diagnosis.xlsx") #   dataset
 
# === Crosstab: counts ===
#    (crosstab)         Diagnosis
#      Thyroid_Cancer_Risk.
crosstab_counts = pd.crosstab(df["Thyroid_Cancer_Risk"], df["Diagnosis"])
print("\n=== Crosstab (Counts) ===") #        .
#        (counts)     .
print(crosstab_counts) 
 
# === Crosstab: row percentages ===
#          (normalize="index"),
#        Diagnosis    .
crosstab_row_pct = pd.crosstab(df["Thyroid_Cancer_Risk"], df["Diagnosis"], normalize="index") * 100
print("\n=== Crosstab (Row %) ===")#        .
print(crosstab_row_pct.round(2)) #       
 
# ===  barplot ===
#      stacked bar chart      
#      .
crosstab_counts.plot(kind="bar", stacked=True, figsize=(8,6))

#         .
plt.title("Distribution of Diagnosis within Thyroid Cancer Risk") 
plt.ylabel("Count") #   y
plt.xlabel("Thyroid Cancer Risk") #   x
plt.legend(title="Diagnosis") #       Diagnosis   bar.
plt.tight_layout() #     
plt.show() #     

#            .
crosstab_row_pct.plot(kind="bar", stacked=True, figsize=(8,6), colormap="coolwarm")

#         () .
plt.title("Row-normalized distribution (%)") 
plt.ylabel("Percentage") #  y
plt.xlabel("Thyroid Cancer Risk") #  x
plt.legend(title="Diagnosis")#       Diagnosis   bar.
plt.tight_layout() #     
plt.show()  #     
```
