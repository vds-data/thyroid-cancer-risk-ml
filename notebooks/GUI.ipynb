```python
import tkinter as tk # # Εισαγωγή τπυ Module (Αρθρώματος) Γραφικών της Python 
from PIL import Image, ImageTk #Για προβολή εικόνων μέσα από την tkinter
from tkinter import ttk #Για τις ανάγκες χρήσης scrollbar

# === Εμφάνιση εικόνας με το Heatmap συσχετίσεων ===
def open_correlation_image():
    # Δημιουργία νέου παραθύρου
    newWindow = tk.Toplevel(mainWindow)
    newWindow.title("Correlation HeatMap")  #Εισαγωγή τίτλου παραθύρου
    newWindow.geometry("800x600") # Ορισμός μεγέθους παραθύρου

    # Δημιουργία πλαισίου (Frame) που θα κρατάει το canvas 
    frame = ttk.Frame(newWindow)
    frame.pack(fill="both", expand=True) #To Frame πιάνει όλο το πλαίσιο
    
    # Canvas  όπου θα τοποθετηθεί η εικόνα
    canvas = tk.Canvas(frame, bg="white")
    canvas.pack(side="left", fill="both", expand=True)

    # Φόρτωση και προβολή εικόνας
    corr_image = Image.open("Correlation/Correlations_for_Continuous_Features.png")
 
    # Zoom factor για μείωση μεγέθους στο 90%
    zoom_factor = .90 
    new_width = int(corr_image.width * zoom_factor) #Νέο πλάτος 
    new_height = int(corr_image.height * zoom_factor) #Νέο ύψος
    corr_image = corr_image.resize((new_width, new_height), Image.LANCZOS) # Επαναδειγματοληψία υψηλής ποιότητας
 
    corrImg = ImageTk.PhotoImage(corr_image) # Μετατροπή σε PhotoImage
 
    # === Κεντράρισμα εικόνας στο canvas ===
    def center_image(event=None):  # Callback όταν αλλάζει το μέγεθος του canvas
        canvas.delete("all") # Καθαρισμός προηγούμενων σχεδίων
        win_width = canvas.winfo_width() # Τρέχον πλάτος canvas
        win_height = canvas.winfo_height() # Τρέχον ύψος canvas
        center_x = win_width // 2 # Κέντρο X
        center_y = win_height // 2 # Κέντρο Υ
        canvas.create_image(center_x, center_y, image=corrImg, anchor="center") # Σχεδίαση εικόνας στο κέντρο
 
    center_image() # Αρχική κλήση για κεντράρισμα
    canvas.bind("<Configure>", center_image) # Re-center σε αλλαγές μεγέθους


# === Εμφάνιση αποτελεσμάτων K-Means ===
def open_k_means_image():
    newWindow = tk.Toplevel(mainWindow) # Δημιουργία νέου παραθύρου
    newWindow.title("K-means Results") #Εισαγωγή τίτλου παραθύρου
    newWindow.geometry("800x600")  # Ορισμός μεγέθους παραθύρου
 
    # Δημιουργία πλαισίου (Frame) που θα κρατάει το canvas και τα scrollbars (εικόνα και τις μπάρες κύλισης)
    frame = ttk.Frame(newWindow)  
    frame.pack(fill="both", expand=True) # Το πλαίσιο πιάνει όλο τον χώρο
 
    # Canvas  όπου θα τοποθετηθεί η εικόνα
    canvas = tk.Canvas(frame) 
    canvas.pack(side="left", fill="both", expand=True)
 
    # Δημιουργία scrollbars (κάθετη και οριζόντια μπάρα κύλισης)
    
    v_scroll = ttk.Scrollbar(frame, orient="vertical", command=canvas.yview)
    v_scroll.pack(side="right", fill="y")
 
    h_scroll = ttk.Scrollbar(newWindow, orient="horizontal", command=canvas.xview)
    h_scroll.pack(side="bottom", fill="x")

    # Συνδέει τις μπάρες κύλισης με την εικόνα
    canvas.configure(yscrollcommand=v_scroll.set, xscrollcommand=h_scroll.set)
    canvas.bind("<Configure>", lambda e: canvas.configure(scrollregion=canvas.bbox("all")))
 
    # Φόρτωση και προβολή εικόνας
    kmeans_image = Image.open("K_Means/K_Means.png")
    kmeansImg = ImageTk.PhotoImage(kmeans_image)
 
    # Δημιουργία Label (πλαισίου) με εικόνα
    kmeansImgLabel = tk.Label(canvas, image=kmeansImg)
    kmeansImgLabel.image = kmeansImg  #Κρατά την εικόνα στη μνήμηγια να μην διαγραφεί από garbage collector 
 
    # Τοποθέτηση εικόνας μέσα στο canvas
    canvas.create_window((0, 0), window=kmeansImgLabel, anchor="nw") 

  # === Εμφάνιση γραφήματων Logistic Regression για Best 5 Triads ===
  
def open_logistic_regression():
    # Δημιουργία νέου παραθύρου
    newWindow = tk.Toplevel(mainWindow)
    newWindow.title("Logistic Regression with Sub-sampling stratified K-fold")  #Εισαγωγή τίτλου παραθύρου
    newWindow.geometry("800x600") # Ορισμός μεγέθους παραθύρου

    # Δημιουργία πλαισίου (Frame) που θα κρατάει το canvas 
    frame = ttk.Frame(newWindow)
    frame.pack(fill="both", expand=True) #To Frame πιάνει όλο το πλαίσιο
    
    # Canvas  όπου θα τοποθετηθεί η εικόνα
    canvas = tk.Canvas(frame, bg="white")
    canvas.pack(side="left", fill="both", expand=True)

    # Φόρτωση και προβολή εικόνας
    LR_image = Image.open("Ordered Barplots _Top_five Trials_LR/Ordered Barplots _Top_five Trials_LR_Merged.png")
 
    # Zoom factor για μείωση μεγέθους στο 17%
    zoom_factor = .17 
    new_width = int(LR_image.width * zoom_factor) #Νέο πλάτος 
    new_height = int(LR_image.height * zoom_factor) #Νέο ύψος
    LR_image = LR_image.resize((new_width, new_height), Image.LANCZOS) # Επαναδειγματοληψία υψηλής ποιότητας
 
    LRImg = ImageTk.PhotoImage(LR_image) # Μετατροπή σε PhotoImage
 
    # Κεντράρισμα εικόνας στο canvas
    def center_image(event=None):  # Callback όταν αλλάζει το μέγεθος του canvas
        canvas.delete("all") # Καθαρισμός προηγούμενων σχεδίων
        win_width = canvas.winfo_width() # Τρέχον πλάτος canvas
        win_height = canvas.winfo_height() # Τρέχον ύψος canvas
        center_x = win_width // 2 # Κέντρο X
        center_y = win_height // 2 # Κέντρο Υ
        canvas.create_image(center_x, center_y, image=LRImg, anchor="center") # Σχεδίαση εικόνας στο κέντρο
 
    center_image() # Αρχική κλήση για κεντράρισμα
    canvas.bind("<Configure>", center_image) # Re-center σε αλλαγές μεγέθους


 # === Εμφάνιση γραφήματων Random Forest για Best 5 Triads ===

def open_random_forest():
    # Δημιουργία νέου παραθύρου
    newWindow = tk.Toplevel(mainWindow)
    newWindow.title("Random Forest with Sub-sampling stratified K-fold") #Εισαγωγή Τίτλου παραθύρου
    newWindow.geometry("800x600")  # Ορισμός μεγέθους παραθύρου
 
    # Δημιουργία πλαισίου (Frame) που θα κρατάει το canvas και τα scrollbars
    frame = ttk.Frame(newWindow)
    frame.pack(fill="both", expand=True) #To Frame πιάνει όλο το πλαίσιο

    # Canvas  όπου θα τοποθετηθεί η εικόνα
    canvas = tk.Canvas(frame, bg="white")
    canvas.pack(side="left", fill="both", expand=True)
    
    # Δημιουργία scrollbars (κάθετη και οριζόντια μπάρα κύλισης)
    v_scroll = ttk.Scrollbar(frame, orient="vertical", command=canvas.yview)
    v_scroll.pack(side="right", fill="y")
    h_scroll = ttk.Scrollbar(newWindow, orient="horizontal", command=canvas.xview)
    h_scroll.pack(side="bottom", fill="x")

    # Συνδέει τις μπάρες κύλισης με την εικόνα
    canvas.configure(yscrollcommand=v_scroll.set, xscrollcommand=h_scroll.set)
 
    # Φόρτωση και προβολή εικόνας 
    RF_image = Image.open("Ordered Barplots _Top_five Trials_RF/Ordered Barplots _Top_five Trials_RF_Merged.png")
 
    # Zoom factor για μείωση μεγέθους στο 57%
    zoom_factor = .57
    new_width = int(RF_image.width * zoom_factor) #Νέο Πλάτος
    new_height = int(RF_image.height * zoom_factor) #Νέο ύψος 
    RF_image = RF_image.resize((new_width, new_height), Image.LANCZOS) # Επαναδειγματοληψία υψηλής ποιότητας
 
    RFImg = ImageTk.PhotoImage(RF_image) # Μετατροπή σε PhotoImage
 
    # Κεντράρισμα εικόνας 
    def center_image(event=None): # Callback όταν αλλάζει το μέγεθος του canvas
        canvas.delete("all") # Καθαρισμός προηγούμενων σχεδίων
        win_width = canvas.winfo_width() #Τρέχον πλάτος
        win_height = canvas.winfo_height() #Τρέχον ύψος
        center_x = win_width // 2 # Κέντρο Χ
        center_y = win_height // 2 # Κέντρο Υ
        canvas.create_image(center_x, center_y, image=RFImg, anchor="center") # Σχεδίαση εικόνας στο κέντρο
 
    center_image() # Αρχική κλήση για κεντράρισμα
    canvas.bind("<Configure>", center_image) # Re-center σε αλλαγές μεγέθους


def open_conclusionsButton():
    # Δημιουργία νέου παραθύρου
    newWindow = tk.Toplevel(mainWindow)
    newWindow.title("Συμπεράσματα") #Τίτλος παραθύρου
    newWindow.geometry("900x700") #Ορισμός μεγέθους παραθύρου 
 
    # Δημιουργία πλαισίου (Frame) που θα κρατάει το canvas και τα scrollbars
    frame = ttk.Frame(newWindow)
    frame.pack(fill="both", expand=True) #To Frame πιάνει όλο το πλαίσιο

    #Δημιουργία κάθετης μπάρας κύλισης
    scrollbar = ttk.Scrollbar(frame)
    scrollbar.pack(side="right", fill="y") #Τοποθέτηση scrollbar στη δεξιά πλευρά
 
    # Text widget με μεγαλύτερη γραμματοσειρά και spacing
    text_widget = tk.Text(
        frame,                       #Προσδιορίζει το παράθυρο εμφάνισης του widget
        wrap="word",                 #Αναδίπλωση του κειμένου ανά λέξη
        yscrollcommand=scrollbar.set,  #Σύνδεση του scrollbar με το widget
        font=("Arial", 16),          #Ορισμός γραμματοσειράς και μεγέθους
        spacing1=10,                 #Απόσταση πριν από κάθε παράγραφο
        spacing2=6,                  #Απόσταση μεταξύ γραμμών
        spacing3=10,                 #Απόσταση μετά από κάθε παράγραφο
        padx=20,                     #Εσωτερικό περιθώριο αριστερά-δεξιά
        pady=20                      #Εσωτερικό περιθώριο πάνω-κάτω
    )
    text_widget.pack(fill="both", expand=True) #Το widget πιάνει όλο το διαθέσιμο πλαίσιο
    scrollbar.config(command=text_widget.yview) #Ενεργοποίηση κύλισης για το widget
 
    # Tags για styling
    text_widget.tag_configure("title", font=("Arial", 18, "bold"), justify="center", spacing1=15, spacing3=15) #Ρυθμίσεις μορφοποίησης τίτλου
    text_widget.tag_configure("body", font=("Arial", 16), justify="left", spacing2=8) #Ρυθμίσεις σώματος κειμένου
    text_widget.tag_configure("subtitle", font=("Arial", 18, "bold"), justify="center", spacing1=15, spacing3=15) #Ρυθμίσεις υπότιτλου
 
    # Κείμενο
    title = "Κρίσιμες Τριάδες που οδηγούν σε Μονοπαραγοντικά Συμπεράσματα\n" #Ορισμός τίτλου
    #Κείμενο σώματος πριν τον υπότιτλο
    body_before = """Με βάση τη σύγκριση των δύο μεθοδολογιών, οι ακόλουθες τριάδες συνιστούν τα πιο αξιόπιστα και ερμηνευτικά ευρήματα:
(Family_History, Iodine_Deficiency, Radiation_Exposure) → Επαναλαμβανόμενα κορυφαία τριάδα και στις δύο τεχνικές.
(Family_History, Iodine_Deficiency, T4_Level) → Ισχυρή παρουσία στη Logistic Regression, επαληθεύεται επίσης σε υψηλή θέση από τον RF.
(Family_History, Iodine_Deficiency, Smoking) → Σημαντική στο RF, με παράλληλη υψηλή κατάταξη και στη LR.
 
Συνεπώς, οι παραπάνω τριάδες αναδεικνύονται ως οι πιο σημαντικές, καθώς η συνέπεια των αποτελεσμάτων τους ενισχύει τη θέση τους ως εν δυνάμει προγνωστικοί δείκτες της επιθετικότητας του καρκίνου στο υπό εξέταση dataset.
Εξετάζοντας ποιοι δείκτες επαναλαμβάνονται συστηματικά μέσα στις κορυφαίες τριάδες τόσο με την μέθοδο Logistic Regression όσο και με την Random Forest. Η λογική είναι ότι ένας δείκτης που εμφανίζεται συχνά σε ισχυρούς συνδυασμούς, διαθέτει από μόνος του υψηλή συσχέτιση με την έκβαση (benign vs malignant).\n
"""
 
    subtitle = "Πέντε πιο σημαντικοί μεμονωμένοι δείκτες\n" #Ορισμός υπότιτλου
    
#Κείμενο σώματος μετά τον υπότιτλο
    body_after = """1. Family History (Οικογενειακό Ιστορικό)
   - Παρουσιάζεται σε όλες σχεδόν τις κορυφαίες τριάδες και με τους δύο αλγορίθμους.
   - Η συνεχής του εμφάνιση καταδεικνύει ότι αποτελεί τον πιο καθολικό προγνωστικό παράγοντα. Βιολογικά εύλογο, καθώς η γενετική προδιάθεση είναι καθοριστική για τον καρκίνο του θυρεοειδούς.
 
2. Iodine Deficiency (Ανεπάρκεια Ιωδίου)
   - Συνδυάζεται με όλους σχεδόν τους υπόλοιπους δείκτες (Radiation, T4, Smoking, Obesity).
   - Εμφανίζεται σταθερά στις πρώτες θέσεις και στους δύο αλγορίθμους.
   - Υπογραμμίζει τη σημασία του μεταβολικού/διατροφικού παράγοντα ως καθοριστικού κινδύνου.
 
3. Radiation Exposure (Έκθεση σε Ακτινοβολία)
   - Κατέχει την κορυφαία θέση στην πιο ισχυρή τριάδα (Family History, Iodine Deficiency, Radiation).
   - Επαληθεύεται ταυτόχρονα τόσο στη μέθοδο Logistic Regression όσο και στην Random Forest.
   - Είναι γνωστός, ισχυρά τεκμηριωμένος παράγοντας κινδύνου για καρκινογένεση.
 
4. T4 Level (Επίπεδα T4)
   - Εμφανίζεται κυρίως στη μέθοδο Logistic Regression, σε τριάδες με υψηλό AUC/Accuracy.
   - Αντικατοπτρίζει τη σημασία των βιοχημικών δεικτών θυρεοειδικής λειτουργίας, οι οποίοι ενισχύουν την ιατρική εγκυρότητα της μελέτης.
   - Χρήσιμος διότι παρέχει κλινικό δείκτη μέτρησης που μπορεί να παρακολουθείται εύκολα.
 
5. Smoking (Κάπνισμα)
   - Παρουσιάζεται ιδιαίτερα στην μέθοδο Random Forest, σε συνδυασμούς με Iodine Deficiency και Family History.
   - Εισάγει έναν lifestyle παράγοντα, διευρύνοντας τη διάσταση των κινδύνων πέρα από τα καθαρά γενετικά/βιοχημικά αίτια.
   - Ενισχύει τη δυνατότητα προληπτικών παρεμβάσεων, καθώς είναι τροποποιήσιμος δείκτης.
"""
 
    # Εισαγωγή τίτλου, σώματος και υπότιτλου
    text_widget.insert("1.0", title, "title") #Εισαγωγή τίτλου στο widget με τη μορφοποίηση "title"
    text_widget.insert("end", body_before, "body") #Εισαγωγή του πρώτου σώματος κειμένου με τη μορφοποίηση "body"
    text_widget.insert("end", subtitle, "subtitle") #Εισαγωγή υπότιτλου με τη μορφοποίηση "subtitle"
    text_widget.insert("end", body_after, "body") #Εισαγωγή του δεύτερου σώματος κειμένου με τη μορφοποίηση "body"
 
    # Read-only
    text_widget.config(state="disabled") #Μετατρέπει το widget σε μόνο για ανάγνωση (δεν επιτρέπει επεξεργασία)
    
#Δημιουργία Κεντρικού Παραθύρου

mainWindow =tk.Tk() #Ορίζουμε μεταβλητή για το κεντρικό παράθυρο

mainWindow.title("Results Visualization - Powered by Vasileia Damaskou") #Τίτλος

#Προβολή Παραθύρου Visualization σε πλήρη οθόνη. 
mainWindow.attributes('-fullscreen',True)

#Δημιουργία Eτικετών (Labels), Πλαισίων Κειμένων (Entries), Κουμπιών (Buttons)
mainLabel = tk.Label(mainWindow, text="Discovering prognostic markers associated with Thyroid Cancer Progression by using Machine Learning tools",
                    font=('Arial',16))

# H πρωτη παραμετρος δείχνει σε πιο παράθυρο θα γίνεται η εμφάνιση του Label
#Η δευτερη παραμετρος αφορά το κείμενο του Label. 
#Η επιπλέον παραμετροι σχετίζονται με μορφοποιησεις

mainLabel.pack(pady=20) #20 Pixels αποσταση απο την κορυφή με βάση των άξονα y. 


# === Προσθήκη Κουμπιών ===

#Επεξηγηματικά σχόλια 
#α)H πρωτη παραμετρος δείχνει σε πιο παράθυρο θα γίνεται η εμφάνιση του Label
#β)Η δευτερη παράμετρος αφορά το κείμενο του Label. 
#γ)Tο command ορίζει τι θα συμβεί κατα την επιλογή του κουμπιού.

# Κουμπί για HeatMap συσχετίσεων
correlationButton = tk.Button(mainWindow, text='Correlation HeatMap', command = open_correlation_image)                              
correlationButton.pack() 

# Κουμπί για K-means Clustering
k_means_button = tk.Button(mainWindow, text='K-means Clustering', command =open_k_means_image)                           
k_means_button.pack()

# Κουμπί για Logistic Regression (Top-5 τριάδες)
logisticregressionButton = tk.Button(mainWindow, text='Best triads with Logistic Regression', command= open_logistic_regression)
logisticregressionButton.pack()  

# Κουμπί για Random Forest (Top-5 τριάδες)
randomforestButton = tk.Button(mainWindow, text= 'Best triads with Random Forest', command= open_random_forest)
randomforestButton.pack()

# Κουμπί για Συμπεράσματα
conclusionsButton = tk.Button(mainWindow, text='Conclusive outcome per attribute', command= open_conclusionsButton)
conclusionsButton.pack()

mainWindow.mainloop() # Εκκίνηση βρόχου μηνυμάτων (event loop)
```
